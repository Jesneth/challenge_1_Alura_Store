#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# 1. Debo conocer el ingreso total de cada tienda.
ingreso_total_tienda1 = sum(tienda.Precio)
ingreso_total_tienda2 = sum(tienda2.Precio)
ingreso_total_tienda3 = sum(tienda3.Precio)
ingreso_total_tienda4 = sum(tienda4.Precio)

# Llamo a cada una de las variables para conocer sus respectivos ingresos
print(f'El ingreso total de la tienda 1 es de: ${ingreso_total_tienda1:,.0f}\n El ingreso total de la tienda 2 es de: ${ingreso_total_tienda2:,.0f}\n El ingreso total de la tienda 3 es de: ${ingreso_total_tienda3:,.0f}\n El ingreso total de la tienda 4 es de: ${ingreso_total_tienda4:,.0f}\n')

```

```python
# Para poderlos graficar debo separar los datos
tiendas = ['Tienda 1', 'Tienda 2','Tienda 3','Tienda 4']
valores = [1150880400,1116343500,1098019600,1038375700]


# Creando gráfico de barras
#plt.figure(figsize=(7, 4)) #Determino el tamaño del grafico, 7 Ancho, 4 Alto(en pulgadas)
plt.ylim(1000000000, 1200000000) #Fijo los limites del eje Y(de abajo hacia arriba)
plt.bar(tiendas, valores, color='purple') # plt.bar(eje_x, eje_y, color='opcional') construccion del grafico
plt.title('Ingresos totales por tienda', fontsize=16) #titulo grande por encima del grafico
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ', fontsize = 12)


# Mostar las etiquetas con los valores en cada barra
for i, ingreso in enumerate(valores):
    plt.text(i, ingreso + 5000000, f"${ingreso:,.0f}", ha='center', va='bottom', fontsize=9)

#Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda

tienda1 + categoria.producto

```python
#1.calcular la cantidad de productos vendidos por categoría en cada tienda
categorias_t1 = tienda['Categoría del Producto'].value_counts()
categorias_t2 = tienda2['Categoría del Producto'].value_counts()
categorias_t3 = tienda3['Categoría del Producto'].value_counts()
categorias_t4 = tienda4['Categoría del Producto'].value_counts()

#Imprimo los resultados de la busqueda
print(f'Cantidad de productos vendidos por categoría en la Tienda 1:\n {categorias_t1}\n')
print(f'Cantidad de productos vendidos por categoría en la Tienda 2:\n {categorias_t2}\n')
print(f'Cantidad de productos vendidos por categoría en la Tienda 3:\n {categorias_t3}\n')
print(f'Cantidad de productos vendidos por categoría en la Tienda 4:\n {categorias_t4}\n')
```

Grafico los resultados

```python
categorias_t1 = tienda['Categoría del Producto'].value_counts()
categorias_t2 = tienda2['Categoría del Producto'].value_counts()
categorias_t3 = tienda3['Categoría del Producto'].value_counts()
categorias_t4 = tienda4['Categoría del Producto'].value_counts()


df = pd.DataFrame({
    'Tienda 1': categorias_t1,
    'Tienda 2': categorias_t2,
    'Tienda 3': categorias_t3,
    'Tienda 4': categorias_t4
}).fillna(0)

colores_pastel = ['#B5D6D8', '#E0B0FF', '#FAD02E', '#B5E48C']

ax = df.plot(kind='bar', figsize=(12, 6), width=0.8, color=colores_pastel)


plt.title('Cantidad de productos vendidos por categoría en cada tienda', fontsize=16)
plt.xlabel('Categorías', fontsize=12)
plt.ylabel('Cantidad de productos', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)


for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height() + p.get_y()),
                ha='center', va='center', fontsize=9, color='black')


plt.legend(title='Tiendas', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)


plt.tight_layout()


plt.show()
```

Los productos más vendidos por categoria en cada una de las tiendas son los muebles,siendo la tienda 2 la de menor cantidad vendida, seguidamente de la tienda 1, tienda 4 y siendo la tienda 4 la de las ventas más alta en este misma categoria.

Hallando las categorias mas poulares por cada tienda

```python
#2.Determinar las categorías más populares de cada tienda
categoria_popular_t1 = categorias_t1.idxmax()
cantidad_t1 = categorias_t1.max()

categoria_popular_t2 = categorias_t2.idxmax()
cantidad_t2 = categorias_t2.max()

categoria_popular_t3 = categorias_t3.idxmax()
cantidad_t3 = categorias_t3.max()

categoria_popular_t4 = categorias_t4.idxmax()
cantidad_t4 = categorias_t4.max()

#Imprimo las categorias
print(f'Categoría más popular en la Tienda 1: {categoria_popular_t1} ({cantidad_t1} productos vendidos)')
print(f'Categoría más popular en la Tienda 2: {categoria_popular_t2} ({cantidad_t2} productos vendidos)')
print(f'Categoría más popular en la Tienda 3: {categoria_popular_t3} ({cantidad_t3} productos vendidos)')
print(f'Categoría más popular en la Tienda 4: {categoria_popular_t4} ({cantidad_t4} productos vendidos)')
```

Grafico los resultados


```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
categorias_populares = [categoria_popular_t1, categoria_popular_t2, categoria_popular_t3, categoria_popular_t4]
ventas = [cantidad_t1, cantidad_t2, cantidad_t3, cantidad_t4]


plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, ventas, color='mediumseagreen')


for barra, categoria, cantidad in zip(barras, categorias_populares, ventas):
    altura = barra.get_height()


    plt.text(barra.get_x() + barra.get_width()/2, altura + 1, str(cantidad),
             ha='center', va='bottom', fontsize=10, color='black')


    plt.text(barra.get_x() + barra.get_width()/2, altura + (max(ventas) * 0.05), categoria,
             ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')


plt.title('Categoría más popular por tienda', fontsize=14)
plt.xlabel('Tiendas')
plt.ylabel('Cantidad de productos vendidos')
plt.ylim(0, max(ventas) * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.


```python
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones_promedio = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]


plt.bar(tiendas, calificaciones_promedio, color=['#B5D6D8', '#E0B0FF', '#FAD02E', '#B5E48C'])


plt.title('Calificación promedio de los clientes por tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificación promedio', fontsize=12)


for i, v in enumerate(calificaciones_promedio):
    plt.text(i, v + 0.05, f'{v:.2f}', ha='center', fontsize=10)  # Etiquetas sobre las barras


plt.show()
```

Las calificaciones promedios de los clientes es de un rango de 3.98 - 4.05 de satisfaccion siendo la tienda 3 con mayor calificación de 4.05, lo que indica que hay una alta satisfaccion de los clientes referente a los servicios/productos de Alura Store.

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
productos_t1 = tienda['Producto'].value_counts()
productos_t2 = tienda2['Producto'].value_counts()
productos_t3 = tienda3['Producto'].value_counts()
productos_t4 = tienda4['Producto'].value_counts()

def obtener_productos_extremos(productos):
    return productos.idxmax(), productos.max(), productos.idxmin(), productos.min()

producto_mas_vendido_t1, cant_mas_t1, producto_menos_vendido_t1, cant_menos_t1 = obtener_productos_extremos(productos_t1)
producto_mas_vendido_t2, cant_mas_t2, producto_menos_vendido_t2, cant_menos_t2 = obtener_productos_extremos(productos_t2)
producto_mas_vendido_t3, cant_mas_t3, producto_menos_vendido_t3, cant_menos_t3 = obtener_productos_extremos(productos_t3)
producto_mas_vendido_t4, cant_mas_t4, producto_menos_vendido_t4, cant_menos_t4 = obtener_productos_extremos(productos_t4)


tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
productos_mas = [cant_mas_t1, cant_mas_t2, cant_mas_t3, cant_mas_t4]
productos_menos = [cant_menos_t1, cant_menos_t2, cant_menos_t3, cant_menos_t4]
nombres_mas = [producto_mas_vendido_t1, producto_mas_vendido_t2, producto_mas_vendido_t3, producto_mas_vendido_t4]
nombres_menos = [producto_menos_vendido_t1, producto_menos_vendido_t2, producto_menos_vendido_t3, producto_menos_vendido_t4]

tiendas_data = [productos_t1, productos_t2, productos_t3, productos_t4]
titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

colores_pastel = ['#FFB3BA', '#FFDFBA', '#FFFFBA', '#BAFFC9', '#BAE1FF']


tiendas_data = [productos_t1, productos_t2, productos_t3, productos_t4]
titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, data in enumerate(tiendas_data):
    fila = i // 2
    col = i % 2


    top_productos = data.head(5)

    axs[fila, col].pie(
        top_productos.values,
        labels=top_productos.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=colores_pastel
    )
    axs[fila, col].set_title(f'Productos - {titulos[i]}', fontsize=12)

plt.suptitle('Productos más y menos vendidos por tienda', fontsize=16)
plt.tight_layout()
plt.show()
```

Los productos que menos se han vendido no se encuentran en un porcentaje lejano al porcentaje de los productos mas vendidos.Es decir que haymuy buena rotación de los productos.


# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda

```python
promedio_envio_t1 = tienda['Costo de envío'].mean()
promedio_envio_t2 = tienda2['Costo de envío'].mean()
promedio_envio_t3 = tienda3['Costo de envío'].mean()
promedio_envio_t4 = tienda4['Costo de envío'].mean()


print(f'COSTO PROMEDIO DE ENVÍO POR TIENDA:')
print(f'Tienda 1: ${promedio_envio_t1:,.2f}')
print(f'Tienda 2: ${promedio_envio_t2:,.2f}')
print(f'Tienda 3: ${promedio_envio_t3:,.2f}')
print(f'Tienda 4: ${promedio_envio_t4:,.2f}')
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios_envio = [promedio_envio_t1, promedio_envio_t2, promedio_envio_t3, promedio_envio_t4]


color_linea = '#A8D5BA'
color_puntos = '#F8C8DC'


plt.figure(figsize=(8, 5))
plt.plot(tiendas, promedios_envio, marker='o', color=color_linea, linestyle='-', linewidth=2, markersize=8)


plt.title('Costo Promedio de Envío por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo Promedio de Envío ($)', fontsize=12)


for i, value in enumerate(promedios_envio):
    plt.text(tiendas[i], value + 0.1, f"${value:,.2f}", ha='center', va='bottom', fontsize=10)


plt.tight_layout()
plt.show()
```

En cuanto el costo promedio de envio por tienda es la tienda 1, uno de los factores es por su alta facturación donde es evidente que los productos que mas venden son de la categoria de muebles.Lo que haria que el costo de envio se elevara a comparacion de las demas tiendas.
